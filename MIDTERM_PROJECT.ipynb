### C. Dataset Description

The data for this project included the eating habits and physical activity levels of 498 participants aged between 14 and 61 years, which were used to estimate their obesity levels. A survey was administered through a web-based platform where respondents anonymously answered each question in the related study. The relevant paper provides data about the dietary patterns and physical status of individuals from Colombia, Peru, and Mexico. The public dataset in this study has 17 variables, explained below:

**1. Gender:** categorical variable that shows the biological sex of the individual (male or female).

**2. Age:** numerical variable that shows the individual’s age in years.

**3. Height:** numerical variable that shows the individuals’ height in meters.

**4. Weight:** numerical variable that shows the individuals’ weight in kilograms.

**5. Family history of overweight:** categorical variable that shows if the individual has a family member who is overweight or obese (yes or no).

**6. Frequently consumed high-calorie food (FAVC):** categorical variable that shows if the individual often eats high-calorie food (yes or no).

**7. Frequency of consumption of vegetables (FCVC):** ordinal variable that shows how often the individual eats vegetables (1 = never, 2 = sometimes, 3 = always).

**8. Number of main meals (NCP):** ordinal variable that shows how many main meals the individual has daily (1 = between 1 and 2, 2 = three, 3 = more than three, 4 = no answer).

**9. Consumption of food between meals (CAEC):** ordinal variable that shows how often the individual eats food between meals (1 = no, 2 = sometimes, 3 = frequently, 4 = always).

**10. SMOKE:** categorical variable that shows whether the individual smokes or not (yes or no).

**11. Consumption of water daily (CH2O):** ordinal variable that shows how much water the individual drinks daily (1 = less than a liter, 2 = between 1 and 2 L, 3 = more than 2 L).

**12. Monitor calorie intake (SCC):** categorical variable that shows if the individual keeps track of their caloric intake (yes or no).

**13. Frequency of physical activity (FAF):** ordinal variable that shows how often the individual does physical activity (1 = never, 2 = once or twice a week, 3 = two or three times a week, 4 = four or five times a week).

**14. Time using electronic devices (TUE):** ordinal variable that shows how long the individual uses electronic devices (0 = none, 1 = less than an hour, 2 = between one and three hours, 3 = more than three hours).

**15. Consumption of alcohol (CALC):** ordinal variable that shows how often the individual drinks alcohol (1 = no, 2 = sometimes, 3 = frequently, 4 = always).

**16. Type of transportation used (MTRANS):** categorical variable that shows what kind of transportation the individual uses (automobile, motorbike, bike, public transportation, walking).

**17. Level of obesity according to body mass index (NObesity):** ordinal variable that shows the obesity level of the individual according to their BMI (insufficient weight normal weight, overweight level I, overweight level II, obesity type I, obesity type II, obesity type III). The related attribute was the primary outcome.

### D. Dataset cleaning


##### <span style="color:purple;">We created a function to get information about the dataset, also to know central tendency initial values since it was already splitted for a machine learning model implementation, and we wanted to make sure there were no skewed values. Also, as part of the executionable operations within the explore function, we included the "isnull" function to identify any missing values and sum function to add the total missing values if any.Finally we wanted to visualize numerical and categorical variables so we created if statements to do so.</span>.



```python
import matplotlib.pyplot as plt
import seaborn as sns

def explore_data(data):
    
    # Display basic information about the DataFrame
    print("DataFrame Information:")
    print(data.info())

    # Display summary statistics for numerical columns
    print("\nSummary Statistics:")
    print(data.describe())

    # Display the first few rows of the DataFrame
    print("\nFirst Few Rows:")
    print(data.head())

    # Display the count of missing values in each column
    print("\nMissing Values:")
    print(data.isnull().sum())

    # Visualize the distribution of numerical columns
    print("\nDistribution of Numerical Columns:")
    numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
    for column in numerical_columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(data[column], kde=True)
        plt.title(f'Distribution of {column}')
        plt.show()

    # Visualize the count of categorical values
    print("\nCount of Categorical Values:")
    categorical_columns = data.select_dtypes(include='object').columns
    for column in categorical_columns:
        plt.figure(figsize=(8, 5))
        sns.countplot(data[column])
        plt.title(f'Count of {column}')
        plt.show()



```

##### <span style="color:purple;">Function to round to the nearest whole number</span>.


```python
def round_column_to_whole_number(df, column):
    
    def round_to_whole_number(value):
        if pd.notna(value):
            return round(value)
        else:
            return value

    df[column] = df[column].apply(round_to_whole_number)
    return df
```

##### <span style="color:purple;">Function created to round to a decimal</span>.



```python
def round_column_to_second_decimal(df, column):

    def round_to_second_decimal(value):
        if pd.notna(value):
            return round(value, 2)
        else:
            return value

    df[column] = df[column].apply(round_to_second_decimal)
    return df
```

** 
In our data analysis, we have decided not to drop duplicated rows in the DataFrame. This decision was made considering the specific circumstances of the study. It is possible that two or more individuals participating in the study may share the same values for each attribute captured in the dataset.

Dropping duplicated rows indiscriminately could potentially lead to the loss of valuable information and skew our analysis results. By retaining these duplicated rows, we ensure that our analysis reflects the true nature of the data and accounts for any instances where individuals may exhibit similar characteristics.

Therefore, we have chosen to preserve duplicated rows in the DataFrame to maintain the integrity and representativeness of the dataset for our analysis.

### E. Visualizations

##### <span style="color:purple;">We tried to visualize the means in a 3D graph to understand the general aspects arising from the dataset.</span>.

```python
import matplotlib.pyplot as plt
import numpy as np

# Mean values for attributes
attributes = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
mean_values = [24.315964, 1.701620, 86.586452, 2.423496, 2.687826, 2.014685, 2.370914, 0.664614]

# Create figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Bar positions
x_pos = np.arange(len(attributes))

# Plot bars
ax.bar(x_pos, mean_values, zdir='y')

# Set labels and title
ax.set_xticks(x_pos)
ax.set_xticklabels(attributes)
ax.set_xlabel('Attributes')
ax.set_ylabel('Mean Values')
ax.set_zlabel('Mean')

# Show plot
plt.title('Mean Values of Attributes')
plt.show()
```

##### <span style="color:purple;">We wanted to see the 'Age' mean as a parabola to validate that values were correctly projecting correctly.</span>.

```python
# Mean and standard deviation values for Age attribute
mean_age = 24.315964
std_age = 6.357078

# Define range for x values (age)
x = np.linspace(mean_age - 3 * std_age, mean_age + 3 * std_age, 100)

# Calculate corresponding y values for a parabola
y = (x - mean_age) ** 2

# Plotting
plt.figure(figsize=(8, 6))

# Plot parabola
plt.plot(x, y, color='b', label='Parabola')

# Add mean value marker
plt.scatter(mean_age, 0, color='r', label='Mean Age')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Parabola')
plt.title('Parabola Plot with Mean Age Marker')
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
